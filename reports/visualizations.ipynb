```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from collections import Counter
import os
import pathlib
```

```python
df=pd.read_csv('data/bongo_scam_cleaned.csv')
```

```python
#combine messages
all_text="".join(df['Sms'])

word_cloud=WordCloud(width=700,height=500).generate(all_text)

plt.figure(figsize=(10, 5))
plt.imshow(word_cloud, interpolation="bilinear")
plt.title("Word Cloud")
plt.show()
```

```python
word_freq = Counter(all_text.split())
#count each
top_words = word_freq.most_common(10)
print(top_words)
```

```python
plt.figure(figsize=(10, 5))
plt.bar(*zip(*top_words))
plt.xlabel("word")
plt.ylabel("frequency")
plt.show()
```

Linguistic patterns:

1. The presence of 'hela', 'ela' and 'pesa' as some of the top words show that most scam sms involve texts requesting for money
2. Terms like "need" also show there is often a sense of urgency tied to this messages.
3. Lastly, there is often misspelling and/or broken kiswahili, seen from words like 'ela' and 'iyo'
